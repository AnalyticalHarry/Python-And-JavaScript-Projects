## EXAMPLE 1

```python
def insertion_sort(array): 
    #for loop iterating from the second element since the first element is already considered sorted
    for i in range(1, len(array)):
        # Storing the current element as the key
        key = array[i]
        # Initialize the index for comparison with key
        j = i - 1      
        # Compare the key with elements before it and shift them to the right if they are greater
        # While loop continues until we reach the beginning of the array or find an element smaller than the key
        while j >= 0 and key < array[j]:
            # Shift the element to the right
            array[j + 1] = array[j]  
            # Decrement j to compare with the previous element
            j = j - 1               
        # Place the key in its correct position in the sorted part of the array
        array[j + 1] = key
```

```python
data = [9, 5, 1, 4, 3]
```

```python
insertion_sort(data)
print('Sorted Array in Ascending Order:')
print(data)
```

## EXAMPLE 2

```python
def insertion_sort(array, ascending=True):
    """
    Sorts the given array in ascending or descending order using the insertion sort algorithm.

    Args:
        array (list): The array to be sorted.
        ascending (bool, optional): Specifies whether to sort in ascending (True) or descending (False) order. 
                                   Defaults to True for ascending order.

    Returns:
        None. The array is sorted in-place.

    """
    if ascending:
        # Sorting in ascending order
        for i in range(1, len(array)):
            key = array[i]
            j = i - 1
            # Compare key with elements before it and shift them to the right if they are greater
            while j >= 0 and key < array[j]:
                array[j + 1] = array[j]
                j = j - 1
            array[j + 1] = key
    else:
        # Sorting in descending order
        for i in range(1, len(array)):
            key = array[i]
            j = i - 1
            # Compare key with elements before it and shift them to the right if they are smaller
            while j >= 0 and key > array[j]:
                array[j + 1] = array[j]
                j = j - 1
            array[j + 1] = key
```

```python
arr = [4, 2, 1, 3, 5]
insertion_sort(arr)
print(arr)
```

```python
arr = [4, 2, 1, 3, 5]
insertion_sort(arr, ascending=False)
print(arr) 
```

```python
class insert_sort:
    def __init__(self, array):
        self.array = array
    def ascending(self):
        for i in range(1, len(self.array)):
            key = self.array[i]
            j = i - 1
            while j >= 0 and key < self.array[j]:
                self.array[j + 1] = self.array[j]
                j = j - 1
            self.array[j + 1] = key
        return self.array
    def descending(self):
        for i in range(1, len(self.array)):
            key = self.array[i]
            j = i - 1
            while j >= 0 and key > self.array[j]:
                self.array[j + 1] = self.array[j]
                j = j - 1
            self.array[j + 1] = key
        return self.array
```

```python
def InsertSort(List):
    for i in range(1, len(List)):
        current_number = List[i]
        for j in range(1-1,-1,-1):
            if List[j] > current_number:
                List[j], List[j+1] = List[j+1], List[j]
            else:
                List[j + 1] = current_number
                break 
    return List 
```

```python
result = InsertSort(data)
print('Sorted Array in Ascending Order:')
print(result)
```

```python
for i in range(1, len(df)):
    key = df.iloc[i]['Values']
    j = i - 1
    while j >= 0 and df.iloc[j]['Values'] > key:
        df.iloc[j + 1] = df.iloc[j]
        j = j - 1
    df.iloc[j + 1]['Values'] = key
```
