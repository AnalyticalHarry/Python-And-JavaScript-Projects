## ARRAY

!

 Array is a user-defined class that represents a data structure for storing a collection of elements of a specified size. The Array class allows you to set and retrieve values at specific indices within the array.

```python
# Defining class Array
class Array:
    
    # Initializing the class and setting the size of the array
    def __init__(self, size):
        self.size = size
        self.data = [None] * size

    # Method to set the value at a specific index in the array
    def set_value(self, index, value):
        self.data[index] = value

    # Method to get the value at a specific index in the array
    def get_value(self, index):
        return self.data[index]
```

```python
array = Array(5)
```

```python
array.set_value(0, 1)
array.set_value(1, 2)
array.set_value(2, 3)
array.set_value(3, 4)
array.set_value(4, 5)
```

```python
class Matrix:
    def __init__(self, rows, columns):
        # Store the number of rows in the matrix
        self.rows = rows
        # Store the number of columns in the matrix
        self.columns = columns        
        # Create a 2D list representing the matrix with initial values as 0
        self.data = [[0] * columns for _ in range(rows)]  

    def set_value(self, row, column, value):
        # Set the value at the specified row and column in the matrix
        self.data[row][column] = value           

    def get_value(self, row, column):
         # Get the value at the specified row and column in the matrix
        return self.data[row][column]           

    def print_matrix(self):
         # Iterate over each row in the matrix
        for row in self.data: 
            # Iterate over each element in the row
            for element in row:
                 # Print the element followed by a space
                print(element, end=' ')  
                # Print a new line after each row
            print()                               
```

```python
# Create a 3x3 matrix
matrix = Matrix(3, 3)
```

```python
# Set values in the matrix
matrix.set_value(0, 0, 1)
matrix.set_value(0, 1, 2)
matrix.set_value(0, 2, 3)
matrix.set_value(1, 0, 4)
matrix.set_value(1, 1, 5)
matrix.set_value(1, 2, 6)
matrix.set_value(2, 0, 7)
matrix.set_value(2, 1, 8)
matrix.set_value(2, 2, 9)
```

```python
print(matrix.get_value(1, 1))
matrix.print_matrix()
```
