# BUBBLE SORT ALGORITHM1

## HEMANT THAPA

```python
data = [1,3,4,5,2,-7,-5,6,9,-10,10]
```

```python
def bubble_sort(data):
    index_length = len(data) - 1
    sort = False
    
    while not sort:
        sort = True
        for i in range(0, index_length):
            if data[i] > data[i+1]:  
                sort = False
                data[i], data[i+1] = data[i+1], data[i]  
    return data
```

```python
print(bubble_sort(data))
```

```python
def bubble_sort_2(data):
    for i in range(len(data)):
        sort = False
        for j in range(0, len(data) - i - 1):
            if data[j] > data[j + 1]:
                sort = True
                data[j], data[j + 1] = data[j + 1], data[j]
        if not sort:
            break
    return data
```

```python
print(bubble_sort_2(data))
```

```python
def bubble_sort_3(data):
    n = len(data)
    
    for i in range(n - 1):
        sorted = True
        for j in range(n - 1 - i):
            if data[j] > data[j + 1]:
                sorted = False
                data[j], data[j + 1] = data[j + 1], data[j]
        if sorted:
            break
    return data
```

```python
print(bubble_sort_3(data))
```

### OVERALL CODE

#### Ascending

```python
# total length of the list
n = len(data)

# for loop to iterate inside n, which is basically the length of the list data
for i in range(n):
    # Flag to track if any swaps are made in this iteration
    sort = False
    
    # Nested for loop to iterate over the list elements
    # The range is from 0 to n - i - 1, as the largest elements are already sorted in each iteration
    for j in range(0, n - i - 1):
        # Compare adjacent elements and swap them if necessary
        if data[j] > data[j+1]:
            sort = True
            data[j], data[j+1] = data[j+1], data[j]
    
    # If no swaps were made in this iteration, the list is already sorted
    if not sort:
        break
```

```python
print(data)
```

#### Descending

```python
# total length of the list
n = len(data)

# for loop to iterate inside n, which is basically the length of the list data
for i in range(n):
    # Flag to track if any swaps are made in this iteration
    sort = False
    
    # Nested for loop to iterate over the list elements
    # The range is from 0 to n - i - 1, as the largest elements are already sorted in each iteration
    for j in range(0, n - i - 1):
        # Compare adjacent elements and swap them if necessary for descending order
        if data[j] < data[j+1]:
            sort = True
            data[j], data[j+1] = data[j+1], data[j]
    
    # If no swaps were made in this iteration, the list is already sorted
    if not sort:
        break
```

```python
print(data)
```

### FINAL ALGORITHM

```python
def bubble_sort(data, ascending=True):
    n = len(data)
    
    for i in range(n - 1):
        sorted = True
        for j in range(n - 1 - i):
            if ascending:
                # Sort in ascending order
                if data[j] > data[j + 1]:
                    sorted = False
                    data[j], data[j + 1] = data[j + 1], data[j]
            else:
                # Sort in descending order
                if data[j] < data[j + 1]:
                    sorted = False
                    data[j], data[j + 1] = data[j + 1], data[j]
        if sorted:
            break
    return data
```

```python
bubble_sort(data, ascending=True)
```

```python
bubble_sort(data, ascending=False)
```
